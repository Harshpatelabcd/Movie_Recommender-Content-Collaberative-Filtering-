```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import ast
def convert1(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['poster_path'])
        L.append(i['backdrop_path'])
    return L
```

```python
movies['belongs_to_collection']=movies['belongs_to_collection'].apply(convert1)
```

```python
final_movies['genres']= final_movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
final_movies['cast']= final_movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
final_movies['crew']= final_movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])

```

# Text Vectorization (Bags of words)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

Stemming..................

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
import pickle
pickle.dump(similarity,open('content_similarity.pkl','wb'))
```

```python
def recommend(movie_id):
    
    movie_index = new_df[new_df['id'] == movie_id].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:10]
    
    data = []
    
    for i in movies_list:
        item = []
        temp_df = new_df[new_df['id'] == new_df.iloc[i[0]].id]
        item.extend(list(temp_df.drop_duplicates('id')['id'].values))
        item.extend(list(temp_df.drop_duplicates('id')['image1'].values))
        
        data.append(item)
    return data
```

```python
recommend('862')
```

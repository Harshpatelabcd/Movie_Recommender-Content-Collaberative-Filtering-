```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
final_ratings = filtered_rating[filtered_rating['original_title'].isin(famous_movies)]
```

```python
pt = final_ratings.pivot_table(index='original_title',columns='userId',values='rating') 
pt
```

Identifying Uclidian distance of one movies with other 663 movies.....

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

Function that identifies the 6 movies related to movie title which uclidian distance is nearer....

```python
index2 = np.where(pt.index=="X-Men Origins: Wolverine")[0][0]
index2

similar_items1 = sorted(list(enumerate(similarity_scores[index2])),key=lambda x:x[1],reverse=True)[1:6]
similar_items1
```

```python
def recommend(movie_name):
    #index fetch
    index1 = np.where(pt.index==movie_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index1])),key=lambda x:x[1],reverse=True)[1:16]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = movies[movies['original_title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('original_title')['original_title'].values))
        item.extend(list(temp_df.drop_duplicates('original_title')['id'].values))
        item.extend(list(temp_df.drop_duplicates('original_title')['crew'].values))
        
        
        data.append(item)
        
    return data
```
